In this nb we will read de csv and create a dataframe with the one hot encodig for the multiple clases

```python
import pandas as pd
import numpy as np
```

Reading documents in the folder

```python
%%bash
ls ./raw_data
```

```python
df_train = pd.read_csv('./raw_data/train.csv')
df_test = pd.read_csv('./raw_data/test.csv') 
```

## Reading the file to create the dictionary lists

```python
with open('./vocabulary/vocabulary.txt', 'r') as f:
    vocabulary = f.read()
    
vocabulary_list = vocabulary.split('\n')
vocabulary_len = len(vocabulary_list)
```

## Defining the funtion to appen the new columns to the dataframe

```python
def add_dummies(df_row):
    sentences = df_row[-1].split(',')
    
    indexes = []
    for sent in sentences:
        index = vocabulary_list.index(sent)
        indexes.append(index)
        
    onehot_encoded_row = np.zeros(vocabulary_len)
    np.put(onehot_encoded_row, indexes, 1)
    
    return  onehot_encoded_row
        
```

```python
df_train_final = df_train.copy()
df_test_final = df_test.copy()

df_train_final[vocabulary_list] = df_train_final.apply(add_dummies, axis = 1, result_type='expand')
df_test_final[vocabulary_list] = df_test_final.apply(add_dummies, axis = 1, result_type='expand')
```

```python
df_train_final.to_csv('./processed/1_train_dummie.csv', index = False)
df_test_final.to_csv('./processed/1_test_dummie.csv', index = False)
```
